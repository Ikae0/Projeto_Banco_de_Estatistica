Vamos fazer um c√≥digo para verificar se os dados na coluna **Forearm** do dataframe **bf_dataset** seguem uma distribui√ß√£o normal. Utilizamos para tal o teste de normalidade de Shapiro-Wilk. As hip√≥teses s√£o: se o p-valor for maior do que o n√≠vel de signific√¢ncia escolhido, podemos assumir que os dados seguem uma distribui√ß√£o normal. Caso contr√°rio, podemos concluir que os dados n√£o seguem uma distribui√ß√£o normal.
Vamos escolher um valor de n√≠vel de signific√¢ncia como sendo 0.05 por ser o mais comum. Este valor corresponde a uma probabilidade de 5% de cometer um erro do tipo I. Ou seja, de rejeitar a hip√≥tese nula quando ela √© verdadeira.

```python
# teste de normalidade da distribui√ß√£o

from scipy.stats import zscore, norm, shapiro

stat, p_valor = shapiro(bf_dataset['Forearm'])
alpha = 0.05

print(f'O valor estat√≠stico do teste de normalidade de Shapiro-Wilk √©: {stat:.4f}')
if p_valor > alpha:
  print('A distribui√ß√£o n√£o √© normal')
else:
  print('A distribui√ß√£o √© normal')
```

Para a coluna de interesse vamos utilizar algumas fun√ß√µes da biblioteca scipy.stats para calcular o z-score, que √© uma medida de quantos desvios padr√£o um certo valor est√° acima ou abaixo da m√©dia. O z-score √© calculado usando a equa√ß√£o:

$Z = \frac{x - \mu}{\sigma}$\,

sendo $\mu$ a m√©dia e $\sigma$ o desvio padr√£o.

Escolhemos o valor $x = 29$ da distribui√ß√£o para calcular a probabilidade de um valor da coluna **Forearm** ser maior do que esse valor.

Para calcular essa probabilidade utilizamos a seguinte equa√ß√£o para a distribui√ß√£o normal:

$P(X > x) = 1 - P(X \leq x)$\,

onde $P(X \leq x)$ √© calculada usando a distribui√ß√£o acumulada normal para $X$.


```python
# an√°lise dos dados da coluna de interesse (Forearm)

bf_dataset['Z_Score'] = zscore(bf_dataset['Forearm'])

x = 29

probabilidade = 1 - norm.cdf(x, loc = bf_dataset['Forearm'].mean(), scale = bf_dataset['Forearm'].std())

print(f'A probabilidade de Forearm ser maior que {x:.1f} √© de {probabilidade*100:.1f}%')
```

## Quest√£o 4

Escolha uma coluna de valores cont√≠nuos A que seja do seu interesse. Depois disso, filtre os registros da sua tabela usando uma segunda coluna B. E, por fim, valendo-se dos seus conhecimentos sobre p-valor, diga se a m√©dia de A antes da filtragem √© de fato significativamente diferente da m√©dia dos valores de A p√≥s filtragem, com um n√≠vel de confian√ßa de 95%.

> Por exemplo: a coluna BodyFat tem uma determinada m√©dia geral. Se filtrarmos o nosso dataset para que ele inclua apenas registros com circunfer√™ncia abdominal (coluna Abdomen) >= 85, certamente a m√©dia da coluna BodyFat mudar√°. Mas, podemos afirmar, com 95% de confian√ßa estat√≠stica, que essa m√©dia p√≥s-filtragem √©, de fato, diferente da m√©dia original?



---


 A coluna escolhida  √© "Weight" (A) ser√° filtrada a partir da coluna "Knee" (B) determinado pelos valores <=38.

* H0 - hip√≥tese nula: A m√©dia de A ("Weight") se mant√©m igual mesmo ap√≥s a filtragem.
* H1 - hip√≥tese alternativa: A m√©dia de A ("Weight") √© diferente ap√≥s a filtragem.






```python
# M√©dia original da coluna "Weight"
media_peso = bf_dataset['Weight'].mean()
print('M√©dia da coluna A:', media_peso)

std_peso = bf_dataset['Weight'].std()
print("O desvio padr√£o de A antes do filtro:", std_peso, "\n")

# Filtragem
mascara = (bf_dataset['Knee'] <= 38) # Seleciona apenas valores de peso em que o joelho √© <=38

dataset_filtrado = bf_dataset.loc[mascara,['Knee', 'Weight']] # Filtra os valores do peso que est√£o de acordo com a mascara

# M√©dia da coluna "Weight" ap√≥s filtragem
media_peso_filtrado = dataset_filtrado['Weight'].mean()
print("M√©dia da coluna A ap√≥s filtro da coluna B:", media_peso_filtrado)

std_peso_filtrado = dataset_filtrado['Weight'].std()
print("O desvio padr√£o de A ap√≥s o filtro:", std_peso_filtrado)
```

```python
from scipy.stats import ttest_ind

# Realizando teste t de duas amostras
t_stat, p_valor = stats.ttest_ind(bf_dataset['Weight'], dataset_filtrado['Weight'])

# Verificando se o valor p √© menor que o n√≠vel de signific√¢ncia
nivel_confianca = 0.95
alpha = 0.05

if p_valor < (alpha):
    print("Rejeitamos a hip√≥tese nula. A m√©dia ap√≥s o filtro √© significativamente diferente da m√©dia original.")
else:
    print("N√£o rejeitamos a hip√≥tese nula. N√£o h√° evid√™ncia suficiente para afirmar que as m√©dias s√£o diferentes.")

print("Valor t : ", t_stat)
print("Valor p : ", p_valor)
```
